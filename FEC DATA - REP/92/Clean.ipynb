```python
import pandas as pd
```

```python
data = pd.read_csv('bush_92_1.csv')

data['contributor_split'] = data['contributor_employer'].str.split('/')

occ = []
for index, row in data.iterrows():
    if row['contributor_split'] != None:
        if type(row['contributor_split']) != float:
            if len(row['contributor_split']) == 2:
                occ.append(row['contributor_split'][1])
            else:
                occ.append(row['contributor_split'][0])
        else:
            occ.append(row['contributor_split'])
    else:
        occ.append(None)
    

data['contributor_occupation'] = occ

data.to_csv('bush_92.csv')

data.head()
```
