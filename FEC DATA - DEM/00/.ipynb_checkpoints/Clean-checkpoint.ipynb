```python
import pandas as pd
```

```python
gore_00 = pd.read_csv('gore_00_1.csv')

gore_00['contributor_split'] = gore_00['contributor_employer'].str.split('/')

occ = []
for index, row in gore_00.iterrows():
    if row['contributor_split'] != None:
        if type(row['contributor_split']) != float:
            if len(row['contributor_split']) == 2:
                occ.append(row['contributor_split'][1])
            else:
                occ.append(row['contributor_split'][0])
        else:
            occ.append(row['contributor_split'])
    else:
        occ.append(None)
    

gore_00['contributor_occupation'] = occ

gore_00.to_csv('gore_00.csv')

gore_00.head()
```
