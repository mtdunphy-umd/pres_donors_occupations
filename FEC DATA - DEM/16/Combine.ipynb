```python
combine = pd.read_csv("clinton_16_1.csv").sample(frac=.25)
for x in range(2, 6):
    temp = pd.read_csv("clinton_16_" + str(x) + ".csv").sample(frac=.25)
    combine.append(temp, ignore_index=True)
    
combine.to_csv('clinton_16.csv')

combine.head()
```
