```python
import pandas as pd
```

```python
first = pd.read_csv("obama_08_1.csv", low_memory=False).sample(frac=.25)
second = pd.read_csv("obama_08_2.csv", low_memory=False).sample(frac=.25)
third = pd.read_csv("obama_08_3.csv", low_memory=False).sample(frac=.25)
fourth = pd.read_csv("obama_08_4.csv", low_memory=False).sample(frac=.25)
fifth = pd.read_csv("obama_08_5.csv", low_memory=False).sample(frac=.25)
```

```python
combine1 = first.append(second, ignore_index=True)
combine2 = combine1.append(third, ignore_index=True)
combine3 = combine2.append(fourth, ignore_index=True)
combine = combine3.append(fifth, ignore_index=True)

combine.to_csv('obama_08.csv', index = False)

combine.head()
```
